# Data Analysis Notebook

This notebook demonstrates how to connect to PostgreSQL, load the `netflix_final` table into a pandas DataFrame, and create basic visualizations similar to the Streamlit dashboard.

```python
import pandas as pd
import plotly.express as px
from scripts.db_connection import get_engine

# establish connection
engine = get_engine()

# load the data
df = pd.read_sql("SELECT * FROM netflix_final", engine)

df.head()
```

## Basic statistics

Let's look at some summary statistics for numeric columns.

```python
df.describe(include='all')
```

## Distribution of release years

```python
fig_year = px.histogram(df, x='release_year', title='Distribution of Release Years')
fig_year.show()
```

## Top countries producing content

```python
top_countries = (df.groupby('country').size().reset_index(name='count')
                 .sort_values('count', ascending=False).head(10))
fig_countries = px.bar(top_countries, x='country', y='count', title='Top 10 Countries')
fig_countries.show()
```

## Type breakdown (Movie vs TV Show)

```python
fig_type = px.pie(df, names='type', title='Movies vs TV Shows')
fig_type.show()
```
