## Handling Missing Values

We'll create a DataFrame containing some `NaN` / null values and
explore various pandas methods such as `.count()`, `.unique()`,
`.sum()`, `.isna()`, and applying custom functions.


```python
import numpy as np

# DataFrame with missing values
students_na = pd.DataFrame({
    "student_id": [1, 2, 3, 4, np.nan],
    "student_name": ["Alice", None, "Charlie", "Diana", "Eve"],
    "age": [20, 22, np.nan, 21, 23]
})

students_na
```

```python
# Methods demonstration
print("Count of non-null values:\n", students_na.count(), "\n")
```

```python
print("Unique values (name):\n", students_na['student_name'].unique(), "\n")
```

```python
print("Sum of student_id column (NaN treated as 0):\n", students_na['student_id'].sum(), "\n")
```

```python
print("Isna matrix:\n", students_na.isna(), "\n")
```

```python
# apply function: add 10 to age if not null
print("\nApply function to age column:")
students_na['age_plus_10'] = students_na['age'].apply(lambda x: x + 10 if pd.notna(x) else x)
students_na
```

## Working with DateTime

DateTime handling is crucial for data analysis. We'll explore:
- Creating datetime objects
- Parsing dates from strings
- DateTime operations in pandas DataFrames
- Formatting and extracting components


```python
import pandas as pd
from datetime import datetime, timedelta

# Create datetime objects
now = datetime.now()
print("Current datetime:", now)
```

```python
# Parse dates from strings
date_str = "2023-10-15"
parsed_date = pd.to_datetime(date_str)
print("Parsed date:", parsed_date)
```

```python
# DataFrame with datetime column
events = pd.DataFrame({
    "event": ["Start", "Mid", "End"],
    "date": pd.to_datetime(["2023-01-01", "2023-06-15", "2023-12-31"])
})

print("\nEvents DataFrame:")
print(events)
```

```python
# Date operations
events['days_since_start'] = (events['date'] - events['date'].min()).dt.days
events['month'] = events['date'].dt.month
events['year'] = events['date'].dt.year

print("\nWith additional columns:")
print(events)
```

```python

# Add days to a date
future_date = parsed_date + timedelta(days=30)
print(f"\n30 days after {parsed_date.date()}: {future_date.date()}")
```

## Understanding the `.apply()` Method

The `.apply()` method in pandas allows you to apply a function along an axis of the DataFrame (rows or columns). It's powerful for:
- Transforming data element-wise
- Performing calculations across rows/columns
- Applying custom logic to your data

Key parameters:
- `func`: The function to apply
- `axis`: 0 for columns, 1 for rows
- Can use lambda functions or named functions


```python
import pandas as pd
import numpy as np

# Sample DataFrame
df = pd.DataFrame({
    'A': [1, 2, 3, 4],
    'B': [10, 20, 30, 40],
    'C': ['a', 'b', 'c', 'd']
})

print("Original DataFrame:")
df
```

```python
# Apply to columns (axis=0) - sum each column
print("\nSum of each column:")
print(df[['A', 'B']].apply(np.sum, axis=0))
```

```python
# Apply to rows (axis=1) - sum each row
print("\nSum of each row:")
df[['A', 'B']].apply(np.sum, axis=1)
```

```python
# Using lambda function - double values in column A
print("\nDouble values in column A:")
df['A_doubled'] = df['A'].apply(lambda x: x * 2)
df
```

```python
# Custom function - categorize ages
def categorize_age(age):
    if pd.isna(age):
        return 'Unknown'
    elif age < 20:
        return 'Teen'
    elif age < 25:
        return 'Young Adult'
    else:
        return 'Adult'

# Apply to the students_na DataFrame from earlier
students_na['age_category'] = students_na['age'].apply(categorize_age)
print("\nStudents with age categories:")
students_na[['student_name', 'age', 'age_category']]
```
